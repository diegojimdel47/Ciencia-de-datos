```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

try:
  df = pd.read_csv('train.csv')
  print("Dataset cargado exitosamente")
except FileNotFoundError:
  print("Error: No se encontró el archivo .csv")

df.head()
```

```python
df.columns = df.columns.str.strip().str.lower()

print("Columnas disponibles:\n")
print(df.columns)

print("\nInformación general del dataset: \n")
print(df.info())

print("\nEstadísticas descriptivas generales:\n")
print(df.describe())

posibleVar = ['log_price', 'accommodates', 'bathrooms', 'bedrooms', 'beds']
keyVar = [col for col in posibleVar if col in df.columns]

print("\nMedia:\n")
print(df[keyVar].mean())

print("\nMediana:\n")
print(df[keyVar].median())

print("\nDesviación estándar:\n")
print(df[keyVar].std())

print("\nVarianza:\n")
print(df[keyVar].var())

print("\nRango:\n")
print(df[keyVar].max() - df[keyVar].min())

print("\nValores nulos:\n")
print(df[keyVar].isnull().sum())

print("Matriz de correlación:\n")
print(df[keyVar].corr())
```

```python
plt.figure()
sns.histplot(df['log_price'], kde=True)
plt.title("Distribución del precio")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='room_type', y='log_price', data=df)
plt.title("Precio según tipo de habitación")
plt.xlabel("Tipo de habitación")
plt.ylabel("Precio")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='accommodates', y='log_price', data=df)
plt.title("Relación entre capacidad y precio")
plt.xlabel("Capacidad")
plt.ylabel("Precio")
plt.show()
```

```python
plt.figure()
sns.countplot(
    y='property_type',
    data=df,
    order=df['property_type'].value_counts().index[:10]
)
plt.title("Tipos de propiedades más comunes")
plt.show()
```

```python
plt.figure()
sns.heatmap(df[['log_price', 'accommodates', 'bathrooms', 'bedrooms', 'beds']].corr(),
            annot=True, cmap='coolwarm')
plt.title("Matriz de correlación")
plt.show()
```
