# Ejercicio 1

```python
# Formula para la normalización
# X_normalized = (X - X_min) / (X_max - X_min)
data = [10, 20, 30, 40, 50]
x_min = min(data)
x_max = max(data)

n1 = (data[0] - x_min) / (x_max - x_min)
n2 = (data[1] - x_min) / (x_max - x_min)
n3 = (data[2] - x_min) / (x_max - x_min)
n4 = (data[3] - x_min) / (x_max - x_min)
n5 = (data[4] - x_min) / (x_max - x_min)

normalized_data = [n1, n2, n3, n4, n5]

print("Datos base:", data)
print("Datos normalizados:", normalized_data)
```

# Ejercicio 2

```python
# Formula de estandarización
# Z = (X - μ) / σ
import statistics

data = [2, 4, 4, 4, 5, 5, 7, 9]
mu = statistics.mean(data)
sigma = statistics.stdev(data)

z1 = (data[0] - mu) / sigma
z2 = (data[1] - mu) / sigma
z3 = (data[2] - mu) / sigma
z4 = (data[3] - mu) / sigma
z5 = (data[4] - mu) / sigma
z6 = (data[5] - mu) / sigma
z7 = (data[6] - mu) / sigma
z8 = (data[7] - mu) / sigma

standarized_data = [z1, z2, z3, z4, z5, z6, z7, z8]

print("Datos base:", data)
print("Datos estandarizados:" , standarized_data)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [np.nan, 2, 3, 4, np.nan],
    'C': [1, 2, 3, 4, 5]
})

print("Valores nulos:")
print(df.isnull())
print("\n")

print("Valores nulos por columna")
print(df.isnull().sum())
print("\n")

print("Porcentaje de valores nulos:")
print(df.isnull().mean() * 100)
print("\n")

print("Filas con valores nulos:")
print(df[df.isnull().any(axis=1)])
```

# Ejercicio 5

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [np.nan, 2, 3, 4, np.nan],
    'C': [1, 2, 3, 4, 5]
})

print("Filas sin valores nulos:")
df_drop_rows = df.dropna()
print(df_drop_rows)
print("\n")

print("Columnas sin valores nulos:")
df_drop_cols = df.dropna(axis=1)
print(df_drop_cols)
print("\n")

print("Imputación con media")
df_mean = df.fillna(df.mean())
print(df_mean)
print("\n")

print("Imputación con mediana")
df_median = df.fillna(df.median())
print(df_median)
print("\n")

print("Imputación con Forward Fill")
df_ffill = df.ffill
print(df_ffill)
print("\n")

print("Imputación con bacward fill")
df_bfill = df.bfill
print(df_bfill)
```

# Ejercicio 6

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [np.nan, 2, 3, 4, np.nan],
    'C': [1, 2, 3, 4, 5]
})

print("Imputación con media")
imputer_mean = SimpleImputer(strategy='mean')
df_mean = pd.DataFrame(imputer_mean.fit_transform(df), columns=df.columns)
print(df_mean)
print("\n")

print("Imputación con mediana")
imputer_median = SimpleImputer(strategy='median')
df_median = pd.DataFrame(imputer_median.fit_transform(df), columns=df.columns)
print(df_median)
print("\n")

print("Imputación con most frequent")
imputer_mode = SimpleImputer(strategy='most_frequent')
df_mode = pd.DataFrame(imputer_mode.fit_transform(df), columns=df.columns)
print(df_mode)
print("\n")

print("Imputación con constant")
imputer_const = SimpleImputer(strategy='constant', fill_value=0)
df_const = pd.DataFrame(imputer_const.fit_transform(df), columns=df.columns)
print(df_const)
```

# Ejercicio 7

```python
import numpy as np

data = [10, 12, 14, 15, 16, 18, 20, 22, 25, 100]

Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1
inf_lim = Q1 - 1.5 * IQR
sup_lim = Q3 - 1.5 * IQR
outliers = [x for x in data if x < inf_lim or x > sup_lim]

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Limite inferior:", inf_lim)
print("Limite superior:", sup_lim)
print("Outliers:", outliers)
```

# Ejercicio 8

```python
from scipy import stats
import numpy as np

data = np.array([10, 12, 14, 15, 16, 18, 20, 22, 25, 100])

z_scores = stats.zscore(data)
outlier_index = np.where(np.abs(z_scores) > 3)
outliers = data[outlier_index]

print("Z-Scores:", z_scores)
print("Indices de outliers", outlier_index)
print("Valores de outliers", outliers)
```

# Ejercicio 9

```python
import numpy as np
import pandas as pd
from scipy.stats import boxcox

data = np.array([10, 12, 14, 15, 16, 18, 20, 22, 25, 100])
df = pd.DataFrame(data, columns=["valor"])

Q1 = df["valor"].quantile(0.25)
Q3 = df["valor"].quantile(0.75)
IQR = Q3 - Q1

inf_lim = Q1 - 1.5 * IQR
sup_lim = Q3 + 1.5 * IQR

print("Límite inferior: ", inf_lim)
print("Límite superior", sup_lim)

df_no_outliers = df[(df["valor"] >= inf_lim) & (df["valor"] <= sup_lim)]

df_capping = df.copy()
df_capping["valor"] = np.where(df_capping["valor"] > sup_lim, sup_lim, df_capping["valor"])
df_capping["valor"] = np.where(df_capping["valor"] < inf_lim, inf_lim, df_capping["valor"])

df_log = df.copy()
df_log["valor"] = np.log(df_log["valor"])

df_boxcox = df.copy()
df_boxcox["valor"], lambda_ = boxcox(df_boxcox["valor"])

print("Datos base:\n", df)
print("\n")

print("Sin outliers:\n", df_no_outliers)
print("\n")

print("Capping:\n", df_capping)
print("\n")

print("Log:\n", df_log)
print("\n")

print("Boxcox:\n", df_boxcox)
print("Lambda:\n", lambda_)
```
